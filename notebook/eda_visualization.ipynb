```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter("ignore")
```

```python
df = pd.read_csv("sales_data.csv")
df
```

```python
product_sales = df.groupby("Product")["Sales"].sum()
print(product_sales)
```

```python
plt.figure()
product_sales.plot(kind="bar")
plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
plt.figure()
plt.plot(df["Date"], df["Sales"])
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
# 1. Laptop has the highest total sales compared to Mobile and Tablet.
# 2. East and North regions contribute higher profit on average.
# 3. Sales show fluctuations over time with an overall stable trend.
# 4. Higher sales generally lead to higher profit.
```
